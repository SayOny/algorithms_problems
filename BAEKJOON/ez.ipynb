```python
# 10809 알파벳 찾기

S = input()
alphabet = {chr(i): -1 for i in range(ord('a'), ord('z') + 1)}

for i, s in enumerate(S):
    if alphabet[s] == -1:
        alphabet[s] = i

for value in alphabet.values():
    print(value, end=' ')
```

```python
# 10250 ACM 호텔 

T = int(input())

for _ in range(T):
    H, W, N = map(int, input().split())

    floor = N%H
    ho = (N//H)+1

    if floor == 0:
        floor = H
        ho -= 1
    
    if ho < 10:
        print(str(floor)+"0"+str(ho))
    else:
        print(str(floor)+str(ho))
```

```python
# 4153 직각삼각형

while(True):
    triangle = list(map(int, input().split()))
    if sum(triangle) == 0:
        break

    a = triangle.pop(triangle.index(max(triangle)))
    b = triangle.pop()
    c = triangle.pop()

    if (a**2) == ((b**2)+(c**2)):
        print("right")
    else:
        print("wrong")
```

```python
# 1978 소수 찾기

N = int(input())
ls = list(map(int, input().split()))
answer = 0


for i in ls:
    if i == 1:
        continue

    flag = True

    for j in range(2, i+1):
        if (i%j==0) and (i!=j):
            flag = False

    if flag == True:
        answer += 1

print(answer)
```

```python
# 2231 분해합

def find_smallest_constructor(N):
    for M in range(max(1, N - 54), N):
        if M + sum(map(int, str(M))) == N:
            return M
    return 0

N = int(input())

print(find_smallest_constructor(N))
```

```python
# 2798 블랙잭

from itertools import combinations

def blackjack(N, M, cards):
    max_sum = 0
    
    # 카드에서 3장을 고르는 모든 조합을 확인
    for combo in combinations(cards, 3):
        total = sum(combo)
        if total <= M:
            max_sum = max(max_sum, total)
    
    return max_sum

# 입력 받기
N, M = map(int, input().split())
cards = list(map(int, input().split()))

# 함수 호출
result = blackjack(N, M, cards)
print(result)
```

```python
# 15829 Hashing

def calculate_hash(L, string):
    r = 31
    M = 1234567891
    hash_value = 0
    power_r = 1 

    for i in range(L):
        a_i = ord(string[i]) - ord('a') + 1
        hash_value = (hash_value + a_i * power_r) % M
        power_r = (power_r * r) % M

    return hash_value

L = int(input())  
string = input() 

print(calculate_hash(L, string))
```
