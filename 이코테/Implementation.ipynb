### 상하좌우

```python
# 1: 답

# N 입력받기
n = int(input())
x, y = 1, 1
plans = input().split()

# L, R, U, D에 따른 이동 방향
dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
move_types = ['L', 'R', 'U', 'D']

# 이동 계획을 하나씩 확인
for plan in plans:
    # 이동 후 좌표 구하기
    for i in range(len(move_types)):
        if plan == move_types[i]:
            nx = x + dx[i]
            ny = y + dy[i]
    # 공간을 벗어나는 경우 무시
    if nx < 1 or ny < 1 or nx > n or ny > n:
        continue
    # 이동 수행
    x, y = nx, ny

print(x, y)
```

```python
# 2: 내 첫 풀이

n = int(input())
directions = input().split()
row, col = 1, 1
for d in directions:
    if d == "L" and 1 < col:
        col -= 1
    elif d == "R" and col < n:
        col += 1
    elif d == "U" and 1 < row:
        row -= 1
    elif d == "D" and row < n:
        row += 1
print(row, col)
```

```python
# 3: 내 두번째 풀이

n = int(input())
directions = input().split()
x, y = 1, 1

dx = [0, 0, -1, 1]
dy = [-1, 1, 0, 0]
moves = ["L", "R", "U", "D"]

for d in directions:
    i = moves.index(d)
    nx = x + dx[i]
    ny = y + dy[i]

    if 1 <= nx <= n and 1<= ny <= n:
        x, y = nx, ny

print(x, y)
```

```python
# 지피티가 알려준 딕셔너리 방법

n = int(input())
plans = input().split()

x, y = 1, 1

move = {
    "L": (0, -1),
    "R": (0, 1),
    "U": (-1, 0),
    "D": (1, 0)
}

for p in plans:
    dx, dy = move[p]
    nx = x + dx
    ny = y + dy

    if 1 <= nx <= n and 1 <= ny <= n:
        x, y = nx, ny

print(x, y)
```

### Q08 문자열 재정렬

예를 들어 K1KA5CB7이라는 값이 들어오면 ABCKK13을 출력합니다.

출력 조건: 첫째 줄에 문제에서 요구하는 정답을 출력합니다.

```python
def solution(s):
    length = len(s)
    for l in range(1, len(s)+1):
        answer = []
        prev = s[:l] # 0부터 시작하니까 l로 슬라이싱 해도
        cnt = 1
        for i in range(l, len(s)+1, l):
            p = s[i:i+l]
            if p == prev:
                cnt += 1
            elif p != prev:
                if cnt < 2: answer.append(prev)
                else: answer.append(str(cnt)+prev)
                prev = p
                cnt = 1
        if len(s)%l != 0:
            answer.append(s[-(len(s)%l):])
        
        if length >= len("".join(answer)): 
            length = len("".join(answer))
        # print(answer, l)
    return length
    
# solution("aabbaccc")
solution("ababcdcdababcdcd")
```

```python
# 0: 비어 있음, 1: 사과, 2: 몸
n = int(input())
grid = [([0]*n) for i in range(n)]

k = int(input())

for i in range(k):
    r, c = map(int, input().split())
    grid[r-1][c-1] = 1

l = int(input())
rotations = []
directions = []
for i in range(l):
    x, c = input().split()
    rotations.append(int(x))
    directions.append(c)


time = 0

grid[0][0] = 2
snake = [(0, 0)]
r, c = 0, 0

d = 0
ds = {0:(0, 1), 1:(-1, 0), 2:(0, -1), 3:(1, 0)}

while True:
    time += 1

    nr, nc = (r+ds[d][0]), (c+ds[d][1])
    # print("저는 지금", nr, nc, "로 갑니다", ds[d][0], ds[d][1])
    if 0 <= nr < n and 0 <= nc < n:
        # print(nr, nc)
        now = grid[nr][nc]
        if now == 1:
            # print("사과를 먹고 몸이 늘어나요")
            grid[r][c] = 2
            snake.append((nr, nc))
        elif now == 2:
            # print("으악 몸을 만났다")
            break
        else:
            # print("암것두 없네요 이동하는데 꼬리 짜르기")
            grid[snake[0][0]][snake[0][1]] = 0
            snake = snake[1:]
            snake.append((nr, nc))
            grid[nr][nc] = 2
        r, c = nr, nc
    else:
        # print("벽을 만났다")
        break
    
    if time in rotations:
        if directions[rotations.index(time)] == "L":
            d += 1
            d %= 4 
        else:
            d -= 1
            d %= 4 
        # print("이뜹니다", time, d)
    

print(time)
```
