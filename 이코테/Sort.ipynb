### 위에서 아래로
하나의 수열에는 다양한 수가 존재한다. 이러한 수는 크기에 상관 없이 나열되어 있다. 이 수를 큰 수 부터 작은 수의 순서로 정렬해야 한다. 수열을 내림차순으로 정렬하는 프로그램을 만드시오.


입력 조건: 첫째 줄에 수열에 속해 있는 수의 개수 N이 주어진다. (1<=N<=500)
둘째 줄부터 N+1번째 줄까지 N개의 수가 입력된다. 수의 범위는 1 이상 100,000 이하의 자연수이다.


출력 조건: 입력으로 주어진 수열이 내림차순으로 정렬된 결과를 공백으로 구분하여 출력한다.

```python
n = int(input())
ls = []
for i in range(n):
    ls.append(int(input()))

print(" ".join(map(str, sorted(ls, reverse = True))))
```

### 성적이 낮은 순서로 학생 출력하기

N며의 학생 정보가 있다. 학생 정보는 학생의 이름과 성적으로 구분된다. 각 학생의 이름과 성적 정보가 주어졌으 ㄹ때 성적이 낮은 순서대로 학생의 이름을 출력하는 프로그램을 작성하시오


입력 조건: 첫번째 줄에 학생의 수 입력 (1이상 10만 이하), 두번재 줄부터 N+1 번째 줄에는 학생의 이름을 나타내는 문자열 A와 학생의 성적을 나타내는 정수 B가 공백으로 구분되어 입력된다. 문자열 A의 길이와 학생의 성적은 100 이하의 자연수이다.


출력 조건: 모든 학생의 이름을 성적이 낮은 순서대로 출력한다. 성적이 동일한 학생들의 순서는 자유롭게 출력해도 괜찮다.(가나다 순으로 하려면..? 어떻게 해야 하는지도 고려한다.)

```python
N = int(input())
array = []
for i in range(N):
    a, b = input().split()
    array.append((a, int(b)))
array = sorted(array, key = lambda s: s[1])

for i in range(N):
    print(array[i][0], end=' ')
```

### 두 배열의 원소 교체

두 배열 A, B는 N개의 원소로 구성되어 있으며, 배열의 원소는 모두 자연수이다. 최대 K 번의 바꿔치기 연산을 수행할 수 있는데, 바꿔치기 연산이란 배열 A에 있는 원소 하나와 배열 B에 있는 원소 하나를 골라서 두 원소를 서로 바꾸는 것을 말한다. 최종 목표는 배열 A의 모든 원소의 합이 최대하 되도록 하는 것이다. 


N, K 그리고 배열 A와 B의 정보가 주어졌을 때 최대 K 번의 바꿔치기 연산을 수행하여 만들 수 있는 배열 A의 모든 원소의 합이 최댓값을 출력하는 프로그램을 작성하시오.


예를 들어 n=5, k=3이고 배열 a, b가 다음과 같다고 하자


a = [1, 2, 5, 4, 3], b=[5, 5, 6, 6, 5]


이 경우 다음과 같이 세 번의 연산을 수행할 수 있다.


연산 1: a의 1과 b의 6을 바꾸기
연산 2: a의 2와 b의 6을 바꾸기
연산 3: a의 3과 b의 5를 바꾸기

세 번의 연산 이후 배열 a와 b는 다음과 같이 구성된다.


a=[6, 6, 5, 4, 5]
b=[3, 5, 1, 2, 5]


이 때 배열 a의 모든 원소의 합은 26이며, 이보다 더 합을 크게 만들 수 는 없다. 따라서 이 예시의 정답은 26이다.


입력 조건: 첫번째 줄에 n, k가 공백으로 주어진다. (n은 1이상 10만 이하, k는 0이상 n이하), 두번째 줄에 배열 a의 원소들이 공백으로 구분되어 입력된다 모든 원소는 천만보다 작은 자연수이다. 세번째 줄에 배열 b의 원소들이 공백으로 구분되어 입력된다 모든 원소는 천만 보다 작은 자연수이다.   


출력 조건: 최대 k번의 바꿔치기 연산을 수행하여 만들 수 있는 배열 a의 모든 원소의 합의 최댓값을 출력한다.

```python
# 비효율적인 나의 첫번째 코드
n, k = map(int, input().split())
a = list(map(int, input().split()))
b = list(map(int, input().split()))

for i in range(k):
    tmpA = min(a)
    tmpB = max(b)
    a[a.index(min(a))] = tmpB
    b[b.index(max(b))] = tmpA

print(sum(a))
```

```python
# 백준 회의실 배정 1931

n = int(input())

ls = []
for _ in range(n):
    s, e = map(int, input().split())
    ls.append((s, e))

ls.sort(key=lambda x : (x[1], x[0]))

answer = [ls[0]]
for s, e in ls[1:]:
    if s >= answer[-1][1]:
        answer.append((s, e))
        
print(len(answer))
```

```python
# 백준 로프 2217
# 지피티 터치 13번째 줄

n = int(input())
ls = []
for _ in range(n):
    ls.append(int(input()))

ls.sort()

m = 0
for i, mm in enumerate(ls):
    w = mm*(n-i)
    m = max(w, m)

print(m)
```

```python
# 프로그래머스 실패율 42889
# 혼자 함 너무 뿌듯
# 테스트 22 〉	통과 (2934.01ms, 20MB)

def solution(N, stages):
    stages.sort(reverse=True)
    f = []

    for i in range(1, N+1):
        try:
            stages, stage= stages[:stages.index(i)], stages[stages.index(i):]
            l = len(stage)
            f.append((i, l/(len(stages)+l)))
        except:
            f.append((i, 0))

    f.sort(key=lambda x: x[1], reverse=True)

    answer = []
    for i, j in f:
        answer.append(i)
    return answer

solution(5, [2, 1, 2, 6, 2, 4, 3, 3])
```

```python
# 백준 강의실 배정 11000
# 시간 초과 남 진짜 개같네진짜어떡하냐나

n = int(input())

lectures = []
for i in range(n):
    a, b = map(int, input().split())
    lectures.append((a, b))

lectures.sort(key=lambda x : x[1])

classes = []
while lectures:
    c = [lectures[0]]
    lectures = lectures[1:]
    for s, e in lectures:
        if s >= c[-1][1]:
            c.append((s, e))
            lectures = lectures[1:]
    classes.append(c)
print(len(classes))
```

```python
# 백준 강의실 배정 11000

import heapq

n = int(input())

lectures = []
for i in range(n):
    a, b = map(int, input().split())
    lectures.append((a, b))

lectures.sort(key=lambda x : x[0])


heap = []
c = 1
heapq.heappush(heap, lectures[0][1])
for s, e in lectures[1:]:
    if s >= heap[0]:
        heapq.heappop(heap)
    else:
        c += 1
    heapq.heappush(heap, e)

print(len(heap))
```

```python
# 카드 정렬하기
import heapq

n = int(input())

ls = []
for _ in range(n):
    ls.append(int(input()))
heapq.heapify(ls)

cnt = 0
while len(ls) > 1:
    a = heapq.heappop(ls)
    b = heapq.heappop(ls)
    c = a+b
    cnt += c
    heapq.heappush(ls, c)
    
print(cnt)
```
