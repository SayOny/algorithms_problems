### 큰 수의 법칙

```python
N, M, K = map(int, input().split())
ls = list(map(int, input().split()))

ls.sort()
f, s = ls[N-1], ls[N-2]

answer = 0

while True:
    for i in range(K):
        if M == 0:
            break
        answer += f
        M -= 1
    if M == 0:
        break
    answer += s
    M -= 1

print(answer)    
```

## Q 01 모험가 길드

한 마을에 모험가가 N명 있습니다. 모험가 길드에서는 N명의 모험가를 대상으로 공포도를 측정했는데, 공포도가 높은 모험가는 쉽게 공포를 느껴 위험 상황에서 제대로 대처할 능력이 떨어집니다. 모험가 길드장은 모험가 그룹을 안전하게 구성하고자 공포도가 X인 모험가는 반드시 X명 이상으로 구성항 모험가 그룹에 참여해야 여행을 떠날 수 있도록 규정했습니다. 최대 몇 개의 모험가 그룹을 만들 수 있는지 궁금합니다. 길드장을 위해 N명의 모험가에 대한 정보가 주어졌을 때, 여행을 떠날 수 있는 그룹 수의 최댓 값을 구하는 프로그램을 작성하시오.

예를 들어 N=5이고 각 모험가의 공포도는 2, 3, 1, 2, 2이다.

이때, 그룹 1에 공포도가 1, 2, 3인 모험가를 한 명씩 넣고 그룹 2에 공포도가 2인 남은 두 명을 넣게 되면, 총 2개의 그룹을 만들 수 있습니다. 또한 몇 명의 모험가는 마을에 드래도 남아 있어도 되기 때문에 모든 모허마를 특정한 그룹에 넣을 필요는 없습니다. 

입력 조건: 첫째 줄에 모험가의 수 n이 주어집니다. 1 이상 10만 이하입니다. 둘째 줄에 각 모험가의 공포도의 값을 N 이하의 자연수로 주어지며, 각 자연수는 공백으로 구분합니다.

출력 조건: 여행을 떠날 수 있는 그룹 수의 최댓값을 출력합니다.



```python
from collections import deque

n = int(input())
ls = list(map(int, input().split()))
ls.sort()
ls = deque(ls)
groups = []
while ls:
    group = []
    x = ls[-1]        
    group.append(ls.pop())    
    for i in range(x-1):
        group.append(ls.popleft())
    groups.append(group)
    

print(len(groups))
```

예를 들어 02984라는 문자열이 주어지면 만들어질 수 있는 가장 큰 수는 ((((0+2)X9(X8)X4)=576 입니다. 또한 만들어질 수 있는 가장 큰 수는 항상 20억 이하의 정수가 되도록 입력이 주어집니다.

## Q 03 문자열 뒤집기

다솜이는 0과 1로만 이루어진 문자열 S를 가지고 있습니다. 다솜이는 이 문자열 S에 있는 모든 숫자를 전부 같에 만들려고 합니다. 다솜이가 할 수 있는 행동은 S에서 연속된 하나 이상의 숫자를 잡고 모두 뒤집는 것입니다. 뒤집는 것은 1을 0으로 0을 1로 바꾸는 것을 의미합니다.

예를 들어 S = 0001100일 때는 다음과 같습니다.
1. 전체를 뒤집으면 1110011이 됩니다.
2. 4번재 문자부터 4번째 문자까지 뒤집으면 1111111이 되어서 두 번 만에 모두 같은 숫자로 만들 수 있습니다.

하지만, 처음부터 4번째 문자부터 5번째 문자까지 문자를 뒤집으면 한번에 0000000이 되어서 1번 만에 모두 같은 숫자로 만들 수 있습니다.

문자열 S가 주어졌을 때, 다솜이가 해야 하는 행동의 최소 횟수를 출력하세요.

입력 조건: 첫째 줄에 0과 1로만 이루어진 문자열 S가 주어집니다. S의 길이는 100만보다 작습니다.

출력 조건: 첫째 줄에 다솜이가 해야 하는 행동의 최소 횟수를 출력합니다.


```python
s = list(map(int, input()))

# 1이랑 0으로 다 만들어버리는 거 한 다음에 두 개 비교해서 더 적은 횟수 답으로 내면 되잖슴
def reverse(s, w, d):
     f = 0
     n = 0
     for i, v in enumerate(s):
          if v != w and f == 0:
               f = i
          elif v == d and f != 0:
               f = 0
               n += 1
     return n

print(min(reverse(s, 0, 1), reverse(s, 1, 0)))
```

예를 들어 n=5이고 각 동전이 3, 2, 1, 1, 9원짜리 동전이라고 가정합시다. 이때 만들 수 없는 양의 정수 금액 중 최솟값은 8원입니다.

또 다른 예시로 n=3이고, 각 동전이 각각 3, 5, 7원짜리 동전이라고 가정합시다. 이때 동빈이가 만들 수 없는 양의 정수 금액 중 최솟값은 1원입니다.

입력 조건: 첫째 줄에는 동전의 개수를 나타내는 양의 정수 n이 주어집니다. n은 1이상 1000 이하입니다. 둘째 줄에는 각 동전의 화폐 단위를 나타내는 n개의 자연수가 주어지며 각 자연수는 공백으로 구분합니다. 이때, 각 화폐 단위는 1,000,000 이하의 자연수입니다.

출력 조건: 첫째 줄에 주어진 동전들로 만들 수 없는 양의 정수 금액 중 최솟값을 출력합니다.

```python
n = int(input())
ls = list(map(int, input().split()))
ls.sort()

min_value = 1

# 와 레전드로 어려웠음 진심 개 레전드로 규칙을 몰랐었음. 물론 지금도 떠올리려면 못떠올림
# target~target+num-1 사이 숫자는 이전 범위와 num을 조합하면 무조건 만들어짐
for i in ls:
    if i > min_value:
        break
    min_value += i



print(min_value)
```

## Q 05 볼링공 고르기

A, B 두 사람이 볼링을 치고 있습니다. 두 사람은 서로 무게가 다른 볼링공을 고르려고 합니다. 볼링공은 총 N개가 있으며 각 볼링고마다 무게가 적려 있고 공의 번호는 1번 부터 순서대로 부려됩니다. 또한 같은 무게의 공이 여러 개 있을 수 있지만 서로 다른 공으로 간주합니다. 볼링공의 무게는 1부터 M까지의 자연수 형태로 존재합니다.

예를 들어 N이 5이고, M이 3이며 각각의 무게가 차례대로 1, 3, 2, 3, 2 일 때 각 공의 번호가 차례대로 1번부터 5번까지 부려됩니다. 이 때 두사람이 고를 수 있는 볼링공 번호의 조합을 구하면 다음과 같습니ㅏㄷ.

(1, 2),(1, 3),(1, 4),(1, 5),(2, 3),(2, 5),(3, 4),(4, 5)

결과적으로 두 사람이 공을 고르는 경우의 수는 8가지입니다. N개의 공의 무게가 각가 주어질 때 두 사람이 볼링공을 고르는 경우의 수를 구하는 프로그램을 작성하시오.

입력 조건: 첫째 줄에 볼링공의 개수 N, 공의 최대 무게 M이 공백으로 구분되어 각각 자연수 형태로 주어집니다. n은 1이상 1000이하, m은 1이상 10이하이다. 둘째 줄에 각 볼링공의 무게 k가 공백으로 구분되어 순서대로 자연수 형태로 주어집니다. k는 1이상 m이하.

출력조건: 첫째 줄에 두 사람이 볼링공을 고르는 경우의 수를 출력합니다.

```python
n, m = map(int, input().split())
k = list(map(int, input().split()))
a = 0
for i in range(n):
    for j in range(i+1,n):
        if k[i] != k[j]:
            a += 1
print(a)
```
